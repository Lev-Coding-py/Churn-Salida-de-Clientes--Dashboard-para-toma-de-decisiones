```python
import pandas as pd
```

```python
data=pd.read_csv('customers.csv')
```

```python
data.head()
```

```python
cols = ['SeniorCitizen','tenure','Contract','PaymentMethod',
        'OnlineSecurity','TechSupport','TotalCharges','Churn']
```

```python
data=data[cols]
```

```python
data.head()
```

```python
colum=['Contract','PaymentMethod','OnlineSecurity','TechSupport']

for col in colum:
    print(f'{col}:{data[col].unique()}')
```

```python
data=pd.get_dummies(data,columns=colum,drop_first=False)
```

```python
data.head()
```

```python
z=[]

for col in data:
    if col !='Churn':
        z.append(col)

z.append('Churn')
data=data[z]
        
```

```python
data.head()
```

```python
data['Churn']=data['Churn'].map({'Yes':1, 'No':0})
data['Churn'].unique()
```
