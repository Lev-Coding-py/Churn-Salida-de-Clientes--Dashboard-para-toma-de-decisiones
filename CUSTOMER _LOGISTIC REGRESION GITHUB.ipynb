# PREDICCION E INFLUENCIA DE VARIABLES PREDICTORAS EN CHURN CON REGRESION LOGISTICA
Se utiliza LR para predecir la salida de clientes con una precision del 52%, lo que indica que el modelo puede equivocarse al detectar una salida de clientes. Con un Recall del 85%  que indica que el modelo es casi seguro que predice cuando un cliente se va y no da un falso negativo.

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('customers.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
cols = ['SeniorCitizen','tenure','Contract','PaymentMethod',
        'OnlineSecurity','TechSupport','TotalCharges','Churn']
```

```python
data=data[cols]
```

```python
data.head()
```

```python
d=[]

for col in data.columns:
    if col !='Churn':
        d.append(col)
d.append('Churn')

data=data[d]
```

```python
data.head()
```

```python
data=pd.get_dummies(data,columns=['Contract','PaymentMethod','OnlineSecurity','TechSupport'],drop_first=False)
```

```python
data.info()
```

```python
data.head()
```

```python
data['Churn']=data['Churn'].map({'Yes':1,'No':0})
```

```python
import numpy as np
import pandas as pd

data['TotalCharges'] = data['TotalCharges'].replace(' ', np.nan)
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['TotalCharges'] = data['TotalCharges'].fillna(data['TotalCharges'].median())
```

```python
data['TotalCharges'].info()
```

# LIBRERIA SKLEARN PARA PREDICCION

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

# VARIABLES PREDICTORAS (X) - VARIABLE PREDICTIVA (y)

```python
X=data.drop('Churn',axis=1)
y=data['Churn']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
model = LogisticRegression(max_iter=1000, class_weight="balanced")
model.fit(X_train, y_train)
```

```python
data['Churn'].unique()
```

```python
X.info()
```

# PREDICCION LOGISTICA

```python
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

# MATRIZ DE CONFUSION Y REPORTE DE CLASIFICACION

```python
print(confusion_matrix(y_test,y_predict))
```

```python
print(classification_report(y_test,y_predict))
```

# COEFICIENTES BETA

```python
coeficientes=model.coef_[0]
variables=X.columns

betas=pd.DataFrame({
    'coeficientes':coeficientes,
    'variables':variables
})
```

```python
betas
```
